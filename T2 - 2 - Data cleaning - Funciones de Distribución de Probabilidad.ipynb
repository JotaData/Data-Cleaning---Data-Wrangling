## La simulacion de montecarlo

* Generamos dos numeros aleatorios uniforme x e y entre 0 y 1
* Calcularemos x^2 + y^2
    * Si el valor es inferior a 1 -> estamos dentro del circulo.
    * Si el valor es superior a 1 -> estamos fuera del circulo
* Calculamos el numero total de veces que estan dentro del circulo y lo dividimos entre el numero total de intentos para obtener 
  una aproximacion de la probabilidad de caer dentro del circulo
* Usamos dicha probabilidad para aproximar el valor de pi.
* Repetimos el experimento un numero suficiente de veces (por ej. 100), para obtener (100) diferentes aproximaciones de pi.
* Calculamos el promedio de los 100 experimentos anteriores para dar un valor final de pi.

```python
# Pi comenzara por valer 0, luego creo un array vacio que se ira rellenando
pi_avg = 0
n = 1000
n_exp = 100
pi_value_list = []
for i in range(n_exp):        # range para calcular el experimentos un numero determinado de veces
    value = 0
    x = np.random.uniform(0,1,n).tolist()    # voy a generar el experimento n=1000 veces y luego lo convierto a una lista con tolist()
    y = np.random.uniform(0,1,n).tolist()        
    for j in range(n):      # Ahora un segundo bucle 
        z = np.sqrt(x[j] * x[j] + y[j] * y[j])  # como x e y son listas tengo que acceder a ellos en la posicion j
        if z <= 1:    # si z <= 1 cae dentro del circulo por tanto se suma un nuevo candidato por lo que value va aumentando
            value += 1
    float_value = float(value)   # value hasta la linea anterior es entero, por lo que convierto a float para poder dividir y tener decimales
    pi_value = float_value * 4 / n
    pi_value_list.append(pi_value)  # a pivaluelist le añado con append pi_value
    pi_avg += pi_value  # a pi_avg le puedo sumar +0 el valor de pi, el pi_value que acabamos de calcular

pi = pi_avg/n_exp

print(pi)
plt.plot(pi_value_list)
```

```python
# Para una mejor practica crearemos una definicion "pi_montecarlo"
```

```python
# n y n_exp se han eliminado y han pasado a ser argumentos de la def pi_montecarlo
def pi_montecarlo(n, n_exp):
    pi_avg = 0
    pi_value_list = []
    for i in range(n_exp):
        value = 0
        x = np.random.uniform(0,1,n).tolist()
        y = np.random.uniform(0,1,n).tolist()        
        for j in range(n): 
            z = np.sqrt(x[j] * x[j] + y[j] * y[j]) 
            if z <= 1:
                value += 1
        float_value = float(value)
        pi_value = float_value * 4 / n
        pi_value_list.append(pi_value) 
        pi_avg += pi_value

    pi = pi_avg/n_exp

    print(pi)
    fig = plt.plot(pi_value_list) # guardamos el grafico en fig para usarlo como argumento en "return"
    return(pi, fig)
```

```python
# Ahora usaremos la funcion que hemos definido lineas arriba con los numero de experimentos que deseemos
```

```python
# experimento de tamaño 1000 cada uno de ellos y repetir los experimentos 200 veces
pi_montecarlo(1000, 200)
```
